```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# pick start and end dates
start = '2021-01-01'
end = '2024-12-31'

# pick stocks 
stock1_string = 'ngvc'
stock2_string = 'zion'

# Fetching data for stock1
stock1 = yf.Ticker(stock1_string.lower())
stock1_data = stock1.history(interval='1d',start=start, end=end)
stock1_data['Daily % Change'] = stock1_data['Close'].pct_change().fillna(0)



# Fetching data for stock2
stock2 = yf.Ticker(stock2_string.lower())
stock2_data = stock2.history(interval='1d',start=start, end=end)
stock2_data['Daily % Change'] = stock2_data['Close'].pct_change().fillna(0)


######################################################################### Start plotting #########################################################################
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8))
plt.xticks(rotation=45)

# Plot Daily % Change
ax1.plot(stock1_data['Daily % Change'], color='blue', label=stock1_string.upper())
ax1.plot(stock2_data['Daily % Change'], color='orange', label=stock2_string.upper())
ax1.legend()
ax1.set_title('Daily % Change')
ax1.set_ylabel('Percentage Change')

# Plotting 'Close' with dual y-axes
ax2.plot(stock1_data['Close'], color='blue', label=stock1_string.upper())
ax2.set_ylabel(stock1_string.upper() + ' Price', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

ax3 = ax2.twinx()
ax3.plot(stock2_data['Close'], color='orange', label=stock2_string.upper())
ax3.set_ylabel(stock2_string.upper() + ' Price', color='orange')
ax3.tick_params(axis='y', labelcolor='orange')

# Legend for 'Close'
lines, labels = ax2.get_legend_handles_labels()
lines2, labels2 = ax3.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')
ax2.set_title('Closing Price')

plt.xlabel('Date')
plt.show()

# Calculate stock1 returns over a time period
stock1_start_price = stock1_data['Close'].iloc[0]
stock1_end_price = stock1_data['Close'].iloc[-1]
stock1_percentage_change = ((stock1_end_price - stock1_start_price) / stock1_start_price) * 100

# Calculate stock2 returns over a time period
stock2_start_price = stock2_data['Close'].iloc[0]
stock2_end_price = stock2_data['Close'].iloc[-1]
stock2_percentage_change = ((stock2_end_price - stock2_start_price) / stock2_start_price) * 100

# Reformat both indexes
stock1_data.index = pd.to_datetime(stock1_data.index).strftime('%Y-%m-%d')
stock2_data.index = pd.to_datetime(stock2_data.index).strftime('%Y-%m-%d')

# Calculate correlation
corr = stock1_data['Daily % Change'].corr(stock2_data['Daily % Change'])

################################################################################################################################################################################################
# Print out Stock1 stats
print(f"{stock1_string.upper()} returns since {stock1_data.index[0]}: {round(stock1_percentage_change)}%")
print(f"Avg. Daily Change: {round(stock1_data['Daily % Change'].mean() * 100,3)}%")
print(f"Max Daily Drawback: {round(stock1_data['Daily % Change'].min(),2) * 100}%")
print(f"Max Daily Return: {round(stock1_data['Daily % Change'].max(),2) * 100}%")
print(f"Period Max: ${round(stock1_data['Close'].max(),2)}") 
print(f"Period Low: ${round(stock1_data['Close'].min(),2)}")
print(f"Distance from Period High: ${round(stock1_data['Close'].max() - stock1_data['Close'].iloc[-1],2)} or {round((stock1_data['Close'].max() - stock1_data['Close'].iloc[-1]) / (stock1_data['Close'].max()),2) * 100}% \n")

# Print out Stock2 stats
print(f"{stock2_string.upper()} returns since {stock2_data.index[0]}: {round(stock2_percentage_change)}%")
print(f"Avg. Daily Change: {round(stock2_data['Daily % Change'].mean() * 100,3)}%")
print(f"Max Daily Drawback: {round(stock2_data['Daily % Change'].min(),2) * 100}%")
print(f"Max Daily Return: {round(stock2_data['Daily % Change'].max(),2) * 100}%")
print(f"Period Max: ${round(stock2_data['Close'].max(),2)}") 
print(f"Period Low: ${round(stock2_data['Close'].min(),2)}")
print(f"Distance from Period High: ${round(stock2_data['Close'].max() - stock2_data['Close'].iloc[-1],2)} or {round((stock2_data['Close'].max() - stock2_data['Close'].iloc[-1]) / (stock2_data['Close'].max()),2) * 100}%\n")

print(f"Correlation between {stock1_string.upper()} and {stock2_string.upper()} is {round(corr,3)}")
```
